```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from markovclick.models import MarkovClickstream
from markovclick.viz import visualise_markov_chain
```

```python
df=pd.read_csv("Algorithm data.csv",header=None)
```

```python
df.head()
```

```python
data=[]
for i in range(df.shape[0]):
    x=df.values[i]
    x=x[~pd.isnull(x)]
    data.append(x[1:])
data
```

```python
m=MarkovClickstream(data)
```

```python
fig=plt.figure(figsize=(18,16),dpi=80,facecolor='w',edgecolor='k')
sns.heatmap(m.prob_matrix,xticklabels=m.pages, yticklabels=m.pages, cmap="YlGnBu",annot=True,annot_kws={'size': 15})
```

```python
graph = visualise_markov_chain(m)
```

```python
print(graph.source)
```

```python
graph.render('round-table.gv', view=True)
```
